```python
import numpy as np
import matplotlib.pyplot as plt
# System of equation we are solving here is -
# y - 3*x = 0 and
# -5*x + 2*y = 2
```

```python
# We need sample space (x) to get the corresponding y values for plotting a line.
x_ss = np.linspace(-10, 10, 50)
```

```python
# Let us say first equation as y1 = 3*x and second equation as y2 = 2 + (5*x)/2
y1 = 3 * x_ss
y2 = 1 + (5*x_ss)/2
```

```python
fig, ax = plt. subplots(1, 3)
ax[0].plot(x_ss, y1, label="y = 2x", color="orange")
ax[1].plot(x_ss, y2, label="2y = 2 + 5x", color="yellow")
ax[2].plot(x_ss, y1, label="y = 2x", color="orange")
ax[2].plot(x_ss, y2, label="2y = 2 + 5x", color="yellow")
for axes in ax:
    axes.set_xlim([0, 3])
    axes.set_ylim([3, 8])
    axes.axvline(x=2, color="red", linestyle="--")
    axes.axhline(y=6, color="red", linestyle="--")
plt.tight_layout()
plt.show()
```
